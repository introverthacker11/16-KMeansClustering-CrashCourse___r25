## K Means Clustering for Customer Data


**How K-Means Works:**
- Initialization:
    - Select k (number of clusters to form) and initialize k random centroids (central points of clusters).

- Assignment:
    - Assign each data point to the cluster with the closest centroid (using a distance metric like Euclidean distance).

- Centroid Update:
    - Calculate the mean of all points in each cluster and update the centroid to this mean.

- Repeat:
    - Steps 2 and 3 are repeated until the centroids stabilize (i.e., they no longer change) or a stopping criterion is met (e.g., maximum iterations).

**Applications of K-Means:**
- Customer Segmentation: Grouping customers with similar behaviors.
- Document Clustering: Organizing similar documents together.
- Image Segmentation: Partitioning images into different regions.
- Anomaly Detection: Identifying outliers as separate clusters.
