## Unsupervised Learning
###
Unsupervised learning is a type of machine learning where the algorithm is provided with data that is neither labeled nor categorized. The goal of unsupervised learning is to uncover hidden patterns, structures, or relationships within the data without any explicit guidance or supervision.

Key Characteristics of Unsupervised Learning:

- No Labeled Data: The algorithm works with raw, unstructured data without predefined labels or categories.
- Self-Organization: The algorithm learns to organize the data on its own based on similarities or differences among the data points.
- Exploratory: Often used for exploratory data analysis to discover insights and gain a better understanding of the data.



### Types of Unsupervised Learning
